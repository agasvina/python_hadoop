# Linear Regression 

* Use to fit a line to a data set of observations
* Can be used to predict some value
* Normally calculated using **least square**
* The idea is to minimize the squared-error between each point and the line
* Sometimes called **Maximum likelihood estimation**
* another method to do linear regression is using Gradient Descent
* It iterates to find the line that best follows the contours defined by the data with 3D data


```python
#Initialize the dataset
%matplotlib inline
import numpy as np
from pylab import *

pageSpeeds = np.random.normal(3.0,1.0,1000)
purchaseAmount = 100 - (pageSpeeds + np.random.normal(0,0.1, 1000)) * 3

scatter(pageSpeeds, purchaseAmount)
```

```python
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(pageSpeeds, purchaseAmount)
```

```python
#Checking the R-squared value ~ if the value is close to 1
#it is a good fit
r_value ** 2
```

```python
#Now we construct the equation obtained from the linear regression
import matplotlib.pyplot as plot

def predict(x):
    return slope * x + intercept

fitline = predict(pageSpeeds)

plot.scatter(pageSpeeds, purchaseAmount)
plot.plot(pageSpeeds, fitline, c= 'r')
plot.show()
```
