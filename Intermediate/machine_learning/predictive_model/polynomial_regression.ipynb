# Polynomial Regression

* More often our dataset doesn't look linear
* In this case, we might consider using the polynomial regression
* However, beware of overfitting
* Don't use more degrees than you need

```python
%matplotlib inline
from pylab import *

#Make the random generator deterministic
np.random.seed(2)
pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = np.random.normal(50.0, 10.0, 1000) / pageSpeeds

scatter(pageSpeeds, purchaseAmount)
```

```python
x = np.array(pageSpeeds)
y = np.array(purchaseAmount)

#construct an nth-degree polynomial 
#that minimized squared error
p4 = np.poly1d(np.polyfit(x,y,4))
print np.poly1d(p4)
```

```python
#visualize the predicted value
import matplotlib.pyplot as plot

xp = np.linspace(0,7,100)
plot.scatter(x,y)
plot.plot(xp, p4(xp), c='r')
plot.show()
```

```python
#Measuring the r-squared 
from sklearn.metrics import r2_score
r2 = r2_score(y, p4(x))
print r2
```
